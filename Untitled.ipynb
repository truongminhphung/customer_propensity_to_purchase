# Deal with imbalance datataset and train dataset

### Because this dataset is extremely imbalanced so we will have dataset in some techniques to check what technique is on high accuracy than other

## Technique 1: Oversampling

```python
count_class_0, count_class_1 = df.ordered.value_counts()
print("count_class_0: ", count_class_0)
print("count_class_1: ", count_class_1)
```

```python
# Divide by class
df_class_0 = df[df['ordered'] == 0]
df_class_1 = df[df['ordered'] == 1]
```

```python
df_class_1_over = df_class_1.sample(count_class_0, replace=True)
df_over = pd.concat([df_class_0, df_class_1_over], axis=0)
df_over.shape
```

```python
X = df_over.drop('ordered',axis='columns')
y = df_over['ordered']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=15, stratify=y)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
lr.score(X_test, y_test)
```

```python
y_pred = lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
```

```python
dtc.score(X_test, y_test)
```

```python
y_pred = dtc.predict(X_test)
```

```python
confusion_matrix(y_test, y_pred)
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
```

```python
gnb.score(X_test, y_test)
```

```python
y_pred = gnb.predict(X_test)
confusion_matrix(y_test, y_pred)
```

```python
X = df.drop('ordered',axis='columns')
y = df['ordered']
```

## Technique 2: SMOTE

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_resample(X, y)

y_sm.value_counts()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.2, random_state=15, stratify=y_sm)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(X_train, y_train)
print("lr.score(X_test, y_test): ", lr.score(X_test, y_test))
y_pred = lr.predict(X_test)
print("confusion_matrix(y_test, y_pred): ", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier
dtc = DecisionTreeClassifier()
dtc.fit(X_train, y_train)
print("dtc.score(X_test, y_test): ", dtc.score(X_test, y_test))
y_pred = dtc.predict(X_test)
print("confusion_matrix(y_test, y_pred): ", confusion_matrix(y_test, y_pred))
```

```python
from sklearn.naive_bayes import GaussianNB
gnb = GaussianNB()
gnb.fit(X_train, y_train)
print("gnb.score(X_test, y_test): ", gnb.score(X_test, y_test))
y_pred = gnb.predict(X_test)
print("confusion_matrix(y_test, y_pred): ", confusion_matrix(y_test, y_pred))
```
